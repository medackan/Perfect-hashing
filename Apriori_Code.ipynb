Apriori with Hasing Function

```python
#
itemf=int(input(" item to find:"))
s=10 #Initialize the support value according to your demand.
import pandas as pd
import numpy as np
import time
from mlxtend.frequent_patterns import apriori, association_rules

start_time = time.time()

df = pd.read_csv('4and5data.csv', names = ['products'], sep = ',')
data = df['products'].apply(lambda x:x.split(',') )

init = []
for i in data:
    for q in i:
        if(q not in init):
            init.append(q)
re=[]
for i in range(1,len(init)+1):
    re.append(i)
item_indexx=dict(zip(init,re))
V=[]
for k,v in item_indexx.items():
    V.append(k)
ll = []
for i in data:
    ak = [item_indexx[j] if j in V else j for j in i]
    ll.append(ak)

  



G = []
for i in ll:
    for q in i:
        if(q not in G):
            G.append(q)
from collections import Counter
c = Counter()
my_dicttt={}
for i in G: #printing the frequency of items in dataset
    for d in ll:
        if(i in d):
            c[i]+=1
#print("C1:")
#for i in c:
#    print(str([i])+": "+str(c[i]))
#print()
l = Counter()  #pruning 
for i in c:
    if(c[i] >= s):
        l[frozenset([i])]+=c[i]
for i in l:
    new_keyy =str(list(i))
    new_valuee =l[i]
    my_dicttt[new_keyy] = new_valuee


neww = pd.DataFrame.from_dict(my_dicttt,orient ='index')
neww.rename(columns={ neww.columns[0]: "supp_count" }, inplace = True)
neww.reset_index(level=0, inplace=True)
neww.rename(columns = {'index':'item_sets'}, inplace = True)
D=[]
for i in neww['item_sets']:
    a=eval(i)
    D.append(a[0])
neww['item_sets'] = D

import itertools
b=list(itertools.combinations(neww['item_sets'], itemf))

def c2(n):
    return n*(n-1)/2
def h3(i,j,k,r):
    c=0
    for m in range (r-1, r-i,-1):
        c=c+c2(m)
    return c+h2(j-i,k-i,r-i)
def c3(n):
    return n*(n-1)*(n-2)/6
def h4(i,j,k,l,r):
    c=0
    for m in range (r-1, r-i,-1):
        c=c+c3(m)
    return c+h3(j-i,k-i,l-i,r-i)
def h2(i,j,r):
    return ((i-1)*(2*r-i)/2)+(j-i)
def c4(n):
    return n*(n-1)*(n-2)*(n-3)/24

def h5(i,j,k,l,o,r):
    c=0
    for m in range (r-1, r-i,-1):
        c=c+c4(m)
    return c+h4(j-i,k-i,l-i,o-i,r-i)


r=len(G)
def hashh(e):
    if len(e)==2:
        h=h2(e[0],e[1],r)
    elif len(e)==3:
        h=h3(e[0],e[1],e[2],r)
    elif len(e)==4:
        h=h4(e[0],e[1],e[2],e[3],r)
    elif len(e)==5:
        h=h5(e[0],e[1],e[2],e[3],e[4],r)
    else:
        h=0
    return h

my_dict={}
for e in b:
    new_key = hashh(e) 
    new_value = tuple(e)
    my_dict[new_key] = new_value

A=[]
for i in my_dict.values():
    i=set(i)
    if len(i)==itemf:
        A.append(i)
        
my_dictt = {}
ll_sets = [set(q) for q in ll]  # pre-calculate the sets of ll

# convert A to a tuple of sets to avoid calculating the set of i in each iteration
A_sets = [set(i) for i in A]
for i in A_sets:
    c = 0
    for q_set in ll_sets:
        if i.issubset(q_set):
            c += 1

    new_key = tuple(i)
    new_value = c
    my_dictt[new_key] = new_value

my_dict_filtered = {k: v for k, v in my_dictt.items() if v >= s}


end_time = time.time()
execution_time = end_time - start_time
print(len(my_dict_filtered))#the number of itemset
print("Execution time:", execution_time, "seconds")
```
